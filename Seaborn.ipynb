# Seaborn

O Seaborn é uma biblioteca de visualização de dados do Python baseado no Matplotlib. Ele provê uma interface de alto nível para a construção de gráficos estatísticos atrativos e informativos.

```python
import numpy as np
import pandas as pd
import seaborn as sns

# Definir tema
sns.set_theme(style='darkgrid')
```

```python
# dados
base_dados = sns.load_dataset('tips')

# Ver dados
base_dados.head()
```

```python
# Renomeando colunas
base_dados.rename( columns= {
    'total_bill' : 'Total_Conta',
    'tip' : 'Gorjeta',
    'sex' : 'Sexo',
    'smoker' : 'Fumante',
    'day' : 'Dia_Semana',
    'time' : 'Periodo',
    'size' : 'Tamanho_Mesa'
}, inplace = True)
```

```python
# Gráfico relplot
sns.relplot( x='Total_Conta', y='Gorjeta', data=base_dados );
```

```python
# Gráfico relplot
# Passando outro parâmetro como classe no gráfico
sns.relplot( x="Total_Conta", y="Gorjeta", data=base_dados, hue="Sexo");
```

```python
# Gráfico de linha com 2 eixos
sns.relplot( x='Total_Conta', y='Gorjeta', data=base_dados, kind='line');
```

```python
# Gráfico de linha com 2 eixos
sns.relplot( x="Total_Conta", y="Gorjeta", data=base_dados,
            kind="line", hue="Fumante");
```

```python
# Histograma
sns.histplot( data=base_dados, x="Gorjeta");
```

```python
# Histograma com classe fumante
sns.histplot( data=base_dados, x='Gorjeta', hue='Fumante');
```

```python
# Gráfico de barras
sns.barplot( x='Sexo', y='Total_Conta', data=base_dados, hue='Fumante');
```

```python
# Gráfico Pairplot
sns.pairplot( base_dados );
```

```python
sns.pairplot( data=base_dados, hue = 'Sexo');
```

```python
sns.pairplot( data=base_dados, kind='kde');
```

```python
sns.boxplot( data=base_dados, x='Dia_Semana', y= 'Total_Conta');
```

```python
sns.boxplot( data=base_dados, x='Dia_Semana', y= "Total_Conta", hue='Sexo');
```
