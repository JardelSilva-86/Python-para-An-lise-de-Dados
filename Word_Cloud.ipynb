# O que é Word Cloud

Word Clouds (também conhecidas como wordle, word collage, ou tag word) são representações visuais de palavras que dão maior destaque às palavras que aparecem com mais frequência. No caso das Mentimeter Word Clouds, as palavras adicionadas com mais frequência pelos membros da audiência através de seus smatphones. Esse tipo de visuais ajuda apresentadores  a coletar informações de sua audiência com mais rapidez, destacar as respostas mais comuns e apresentar os dados de uma maneira que todos possam entender.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
# Frases que utilizaremos de exemplo.

frase =  'Lorem ipsum dolor sit amet consectetur adipisicing elit. Distinctio eos laborum molestiae voluptatibus quisquam quos suscipit quaerat, minus minima dolore et dicta libero ex? Iste laborum quo eius labore hic?'
```

```python
# Gerar nuvem de palavra
nuvem_palavra = WordCloud(
    background_color = 'black',
    width = 1600,
    height = 600
).generate( frase )

# Não adianta tentar mostar o objeto com o print. O correto é mostrar em um gráfico
```

```python
# Criando um gráfico

fig, ax = plt.subplots( figsize = (10, 7) )

ax.imshow( nuvem_palavra, interpolation = 'bilinear' )

ax.set_axis_off()
```
