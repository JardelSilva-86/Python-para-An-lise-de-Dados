## Visualização de dados com o Matplotlib

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(7)


# vetor com números inteiros aleatórios entre 1, 1500, com 10 amostras
y = np.random.randint(low = 1, high = 1500, size = 10)

y
```

```python
plt.plot(y)
```

```python
# inserir a primeira linha no plot
plt.plot(y, color = '#749187', marker = 'o', ms = 5, mec = 'k', markerfacecolor = 'w', ls = '-.' )

# inserir a segunda linha no plot
plt.plot(y*2, marker = '+', color = 'k', ms = 5)

# inserir rótulos
plt.xlabel('Eixo X', color = 'red', size = 12)
plt.ylabel('Eixo Y')
plt.title('Título', loc = 'left')

# inserir gridlines
plt.grid(axis = 'y', color = 'gray',
         linestyle = '--', linewidth = 1,
         alpha = 0.8)
plt.show()
```

```python
np.random.seed(6)
x = np.arange(1, 11)
y1 = np.random.randint(1, 400, 10)
y2 = np.random.randint(150, 500, 10)
y3 = np.random.randint(200, 600, 10)

plt.figure(figsize = (15,5))
plt.suptitle("Figura", fontsize = 15)

plt.subplot(1, 3, 1)
plt.plot(x, y1, color = 'k')
plt.title('Subplot 1', pad = 10)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

plt.subplot(1, 3, 2)
plt.plot(x, y2, color = 'r')
plt.title('Subplot 2', pad = 10)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

plt.subplot(1, 3, 3)
plt.plot(x, y3, color = 'g')
plt.title('Subplot 3', pad = 10)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

plt. tight_layout(pad = 4)
plt.show()
```
